```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Generate sample time series data
date_rng = pd.date_range(start='2020-01-01', end='2021-01-01', freq='D')
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randn(len(date_rng))  # Replace with your actual data
df.set_index('date', inplace=True)
```

```python
# Split the data into train and test sets
train_size = int(len(df) * 0.8)
train, test = df[0:train_size], df[train_size:len(df)]

# Fit ARMA model
model = ARIMA(train['data'], order=(2, 0, 2))  # ARMA(p,q) where p=2, q=2
arma_result = model.fit()

# Make predictions
predictions = arma_result.forecast(steps=len(test))
```

```python
# Calculate RMSE and MAE
rmse = np.sqrt(mean_squared_error(test['data'], predictions))
mae = mean_absolute_error(test['data'], predictions)

print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
```
